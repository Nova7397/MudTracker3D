## (Overhang) Edge Angle Detection Script
This script detects overhang angles in a 3D printing image using Hough Line Transform. Detected lines are drawn, and their angles are labeled.

### Requirements
Dependencies: Make sure the following libraries are installed:
    opencv-python
    numpy

### How it Works
    Load Image: Reads the grayscale image and resizes it for processing.

    Detect Lines: Uses the Hough Line Transform to detect lines in the binary image.

    Filter Lines: Removes redundant lines based on a specified angle tolerance (default is ±5°).

Display and Save Results:
Draws detected lines in red.
Annotates the angle of each line in green text.
Display: Shows the final image with detected lines and angles in a new window.

Key Parameters
distance_resolution: Pixel resolution for Hough Line accumulator (1 pixel).
angle_resolution: Angle resolution in radians (1 degree).
threshold: Minimum votes needed to detect a line (default is 40).
Adjust these values to refine line detection as needed.

## Area Calculation
To calculate the total area covered by pure white regions in images, specifically for assessing overhang deviation and zigzag straightness.

### Requirements
Dependencies: Ensure you have OpenCV and NumPy installed.

### Function Details
calculate_white_area(image_path)
Inputs: Path to the image file.

### Process:
Loads the image and creates a mask for pure white pixels ((255, 255, 255)).
Counts the number of white pixels (area).
Calculates the total area of the image.
Computes the percentage of the image that is pure white.
Returns: The white area in pixels and as a percentage of the image.

### Example Output
Running the script displays:

Area of pure white color: [calculated pixels] pixels
Percentage of pure white color: [calculated percentage]%

## Data Visualization
We use this script to visualize prototype ratings data through heatmaps, parallel coordinates, pair plots, and 3D scatter plots, providing insights into various print quality scores like overhang, zigzag, and freeform.

### Requirements
Dependencies: Install the following packages if not already available:
    pandas
    seaborn
    matplotlib

### Plots and Visualizations
1. Heatmaps of Scores
For each score type (Overhang, Zigzag, and Freeform), a heatmap is created.
The heatmaps display how scores vary based on Layer_height and Extrusion_rate.
![Rating assessment/Heatmap.png](Rating assessment/Heatmap.png)

2. Parallel Coordinates Plot
Shows patterns across scores for each combination of Layer_height and Extrusion_rate.
This plot helps in visualizing the relationships and trends across different score types.
![Rating assessment/Parallell Coordinates.png](Rating assessment/Parallell Coordinates.png)

3. Pair Plot
Provides pairwise relationships between all variables in the data.
Useful for identifying correlations and trends between scores and the parameters.
![Rating assessment/Comparison between different factors.png](Rating assessment/Comparison between different factors.png)

4. 3D Scatter Plot
Visualizes the relationship between Layer_height, Extrusion_rate, and the Overhang score.
Offers a spatial perspective to observe how the overhang score is affected by the other two parameters.
